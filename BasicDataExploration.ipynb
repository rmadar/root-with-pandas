```python
# Usual library
import numpy             as np
import pandas            as pd
import matplotlib        as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Plot settings
mpl.rcParams['legend.frameon' ] = False
mpl.rcParams['legend.fontsize'] = 'xx-large'
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16
mpl.rcParams['axes.titlesize' ] = 18
mpl.rcParams['axes.labelsize' ] = 18
mpl.rcParams['lines.linewidth'] = 2.5


# root_pandas package (https://github.com/scikit-hep/root_pandas)
from root_pandas import read_root

# uproot package (https://github.com/scikit-hep/uproot)
import uproot

# root_numpy (http://scikit-hep.org/root_numpy)
from root_numpy import root2array

# pandas_utils (private tool from https://github.com/mickypaganini/YaleATLAS)
import pandautils as pup


def get_pandas_df_uproot(fileName, treeName):
    tfile = uproot.open(fileName)
    tree  = tfile.get(treeName)
    return tree.pandas.df()

def get_pandas_df_numpy_root(fileName, treeName):
    ar = root2array(fileName, treeName).view(np.recarray)
    return pd.DataFrame(ar)

def test_root_file(loadMethod, fileName, treeName=''):
    df = loadMethod(fileName, treeName)
    if ('jet_pt' in df.columns): 
        print('Vector branches seem to be supported\n')
    else:
        print('Vector branches doesn\'t seem to be supported\n')
    %time
    return df
```

```python
%time

# test to create an object with namedtuple (nice but immutable, ie cannot be modified)
import collections
jet = collections.namedtuple('particle', 'pt eta phi')

# Usual dR
def GetdR(j1,j2):
    return np.sqrt( (j1.eta-j2.eta)**2 + (j1.phi-j2.phi)**2 )

# weighted-pT dR computation
def pTweighted_dR(jet1,jet2):
    pt_ave = (jet1.pt + jet2.pt)/2.0
    return pt_ave * np.sqrt( (jet1.eta-jet2.eta)**2 + (jet1.phi-jet2.phi)**2 )

myjet1 = jet(pt=12, eta=0.4, phi=-2.1)
myjet2 = jet(pt=34, eta=0.5, phi=+3.1)
dR     = pTweighted_dR(myjet1,myjet2)
print(myjet1)
print(dR)
```

```python
class Jet(object):
    """
    Jet Class
    """
    def __init__(self, pt = 0 , eta = 0, phi = 0, m = 0):
        super(Jet, self).__init__()
        
        # -- hidden to most usage
        self._pt = pt
        self._eta = eta
        self._phi = phi
        self._m = m
        
    @property
    def pt(self):
        return self._pt
        
    @property
    def eta(self):
        return self._eta

    @property
    def phi(self):
        return self._phi
        
    @property
    def m(self):
        return self._m
```

```python
myJet1 = Jet(pt=35,eta=2.4,phi=-0.4)
myJet2 = Jet(pt=356,eta=0.4,phi=-1.4)
dR     = pTweighted_dR(myJet1,myJet2)
print(dR)
```

```python
%time

# Get a collection of jets from the dataframe information
jets_array = [ Jet(pt, eta, phi) for (pt, eta, phi) in zip(
    pup.flatten(data['jet_pt' ]),
    pup.flatten(data['jet_eta']),
    pup.flatten(data['jet_phi'])
)]

jets_array = np.array( jets_array )
jets = np.array( pup.match_shape( np.array(jets_array) , data['jet_pt'] ) )
```

```python
%time
import itertools

dR_ptweighted=[]
dR=[]
for jet_collection in jets:
    for j in itertools.combinations(jet_collection,2):
        dR.append( GetdR(j[0],j[1]) )
        dR_ptweighted.append( pTweighted_dR(j[0],j[1]) )

plt.figure()
ax = plt.hist(dR,bins=100)
plt.figure()
ax = plt.hist(dR_ptweighted,bins=100)
%time
```

```python
# Now keeps only events for which at exactly 2 leptons are present
data['nlep']  = [len(lep_pt) for lep_pt in data['lep_pt']]
saved_events  = data['nlep']==3
saved_events  = saved_events.values
selected_jets = jets[saved_events]

dR_ptweighted=[]
dR=[]
for jet_collection in selected_jets:
    for j in itertools.combinations(jet_collection,2):
        dR.append( GetdR(j[0],j[1]) )
        dR_ptweighted.append( pTweighted_dR(j[0],j[1]) )

plt.figure()
ax = plt.hist(dR,bins=100)
plt.figure()
ax = plt.hist(dR_ptweighted,bins=100)
```

```python
# generalizing to plot min[dRjj] for different njets values

def plot_mindR_over_event(sample):

    selected_events = sample.index
    selected_jets   = jets[selected_events]
  
    dRmin=[]
    for jet_collection in selected_jets:
        if (len(jet_collection)>1):
            dRmin.append( np.min([GetdR(j[0],j[1]) for j in itertools.combinations(jet_collection,2)]) )
    ax = plt.hist(dRmin,bins=100)    
    return ax


data_per_njets = data.groupby('njets')
plt.figure(figsize=(20,20))
dRmean = {}
for njet,sample in data_per_njets:
    plt.subplot(5,5,njet+1)
    plt.title('$N_{j} = $' + str(njet))
    plt.xlim(0,5)
    ax = plot_mindR_over_event( sample )
    dRmean.update({njet:np.mean(ax[1])})

plt.figure()
plt.xlabel('$N_{j}$')
plt.ylabel('$<dR_{min}>$')
plt.xlim(2,20)
plt.ylim(0,5)
plt.plot(dRmean.keys(),dRmean.values(),'o-')
```
